1) Given,

Two matrices A and B of size nxk and kxm respectively.

Therefore if we multiply two matrices the resultant matrix is of size nxm.

dot product of ith row of A and jth column of B is performed k times.

so the number of multiplications are **nxmxk**.

in dot product of ith row of A and jth column of B the number of additions are k-1.

Hence total number of additions is **nxmx(k-1)**

Given

$$
 f(x) = x^2y+y^3sin(x)
 $$

Gradient of the function with respect to x is:

$$
 \frac{d(f(x,y))}{dx} =
$$

$$
 \frac{d(x^2y)}{dx}+\frac{d(y^3sin(x))}{dx}
$$

$$
 \frac{d(f(x,y))}{dx} = 2xy + y^3cos(x)
$$

Gradient of the function with respect to y is:

$$
 \frac{d(f(x,y))}{dy} =
$$

$$
 \frac{d(x^2y)}{dy}+\frac{d(y^3sin(x))}{dy}
$$

$$
 \frac{d(f(x,y))}{dy} = x^2 + 3y^2sin(x)
$$

```python
#6
import sympy as sp
x,y = sp.symbols('x, y')
gradient_x = sp.diff((x**2)*y + (y**3)*sp.sin(x), x)
gradient_y = sp.diff((x**2)*y + (y**3)*sp.sin(x), y)
gradient = (gradient_x, gradient_y)
print(gradient)
```

```python
#10
import numpy as np
import pandas as pd
matrix = np.random.uniform(low=1, high=2, size=(20, 5))
df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])
std_devs = []
for col in df.columns:
    col_values = df[col].values
    mean = np.mean(col_values)
    variance = np.mean((col_values - mean) ** 2)
    std_dev = np.sqrt(variance)
    std_devs.append((col, std_dev))
# max_std_dev = -float('inf')
column_with_highest_std = None
for col, std_dev in std_devs:
    if std_dev > max_std_dev:
        max_std_dev = std_dev
        column_with_highest_std = (col, std_dev)
means = []
for i in range(df.shape[0]):
    row_values = df.iloc[i].values
    mean = np.mean(row_values)
    means.append((i, mean))
min_mean = float('inf')
row_with_lowest_mean = None
for index, mean in means:
    if mean < min_mean:
        min_mean = mean
        row_with_lowest_mean = (index, mean)
print("DataFrame:\n", df)
print("\nColumn with the highest standard deviation:", column_with_highest_std)
print("\nRow with the lowest mean:", row_with_lowest_mean)
```
